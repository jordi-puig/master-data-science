## 0. El entorno __BlackJack__

L'entorn __BlackJack__ simula una partida de blackjack, que es juga amb una baralla francesa de 52 cartes (baralla de pòquer), constituïda per 4 pals (trèvol ♣, cors ♥, piques ♠, i diamants ♦) i 13 cartes per pal (As , els números del 2 al 10, J, Q i K).

L'objectiu d'aquest joc és obtenir cartes el valor numèric de les quals sumin el nombre més gran possible sense excedir de 21.

Totes aquelles cartes amb un valor numèric compten com a dit valor numèric, les figures (J, Q i K) compten com un 10, i l'as pot comptar com un 1 (en aquest cas s'anomena as no usable) o com un 11 ( en aquest cas es denomina as usable).

A la figura següent es pot observar el valor numèric de tots els tipus de cartes:<br>

BlackJackCards.png

**Figura 1. El joc del Blackjack.** (Font: Sutton & Barto, 1998)

Les normes del joc estan explicades al mòdul 5 de l'assignatura.

Jugar al blackjack pot ser formulat de forma natural com un MDP finit
i episòdic

• Cada partida de blackjack és un episodi.<br>
• Les recompenses són: +1 per guanyar, -1 per perdre i 0 per empatar.<br>
• El retorn es calcula sense descompte (γ = 1).<br>
• El conjunt d'accions del jugador és demanar carta (hit) o ​​plantar-se (stick).
És a dir: A = {hit,stick}.<br>
• Els estats depenen de les cartes del jugador i de la carta que mostra el
crupier.<br>
• El jugador pren decisions basant-se en tres variables:<br>
– Si li ha sortit un as que pugui comptar com a 11 sense passar-se (usable ace a
anglès).<br>
– La suma actual de les cartes (de 12 a 21).<br>
– El valor de la carta de panxa enlaire del crupier.<br>
• Això fa un total de 200 estats diferents.<br>

L'entorn que utilitzarem inicialment implementa totes les característiques comentades anteriorment i la política següent per al crupier:


• Es planta si la suma és 17 o més gran, demana carta en un altre cas.


El codi per implementar aquest entorn està disponible al fitxer adjunt `blackjack_env.py`

Començarem carregant l'entorn i veure quines característiques té, executant un episodi de prova.

### 0.2. Execució d'un episodi

A continuació, realitzarem l'execució d'un episodi de l'entorn `BlackjackEnv` utilitzant un agent que selecciona les accions de forma aleatòria.


```python
# Inicialitzem l'entorn
obs = env.reset()
t, total_reward, done = 0, 0, False

print("Obs inicial: Player = {}, Dealer = {}, Usable Ace = {} ".format(obs[0],obs[1],obs[2]))

switch_action = {
        0: "Stick",
        1: "Hit",
    }

while not done:
    
    # Elegir una acción aleatoria (ésta es la implementación del agente)
    action = env.action_space.sample()
    
    # Ejecutar la acción y esperar la respuesta del entorno
    new_obs, reward, done, info = env.step(action)
    
    # Imprimir time-step
    print("Action: {} -> Obs: Player = {}, Dealer = {}, Usable Ace = {} and reward: {}".format(switch_action[action], new_obs[0],new_obs[1],new_obs[2], reward))
    
    # Actualizar variables
    obs = new_obs
    total_reward += reward
    t += 1
    
print("Episodio finalizado después de {} pasos y la recompensa fue {} ".format(t, total_reward))
env.close()
```
