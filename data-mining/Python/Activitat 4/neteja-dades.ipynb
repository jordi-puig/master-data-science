## Introducció

En aquesta activitat aprendrem a netejar les dades, el pas necessari abans de fer qualsevol anàlisi o model. Això inclou tractar les dades buides o dades que falten, convertir tipus de dades i binaritzar variables.

El conjunt de dades que farem servir en aquesta activitat correspon a les metadades de més de 5000 películes a IMDB, el [IMDB 5000 Movie Dataset](https://data.world/popculture/imdb-5000-movie-dataset).

El primer que farem, com sempre, serà llegir les dades i donar-li un cop d'ull.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/movie_metadata.csv')
df.head()
```

```python
df.describe()
```

## Tractament de les dades no disponibles

Un dels problemes més habituals que ens podem trobar en els nostres conjunts de dades és que no tinguem totes les observacions per tots els registres. És a dir, que algunes de les variables no estiguin disponibles. Aquest fet es pot donar per vàries raons: perquè aquestes dades no existeixen, perquè no es van introduir correctament o simplement per algun error de computació. El cas és que en la majoria d'anàlisis, les dades no disponibles poden causar errors. A continuació veurem tres estratègies diferents per solucionar aquest problema.

- Eliminar les variables amb un percentatge alt de valors buits
- Eliminar les files amb variables no disponibles
- Imputar les dades o omplir-les amb un valor per defecte

### Eliminar les variables amb un percentatge alt de valors buits

En el cas poc habitual que tinguem columnes sense informació, amb tots els valors buits, podem aplicar la funció `dropna` amb el paràmetre `axis=1`, indicant que volem aplicar-lo si es compleix per totes les files, amb `how=all`.

En el nostre cas, no tenim cap columna buida. Per veure-ho, primer en crearem una.

```python
df['missings'] = np.nan
```

Veiem que tenim una nova columna anomenada `missings`, amb tots els valors nuls.

```python
'missings' in df.columns and df['missings'].isnull().all()
```

Eliminem les columnes amb tots els valors `NaN`. El paràmetre `inplace=True` indica que l'operació modificarà el dataset.

```python
df.dropna(axis=1, how='all', inplace=True)
```

Ara ja no tenim la columna al _dataset_.

```python
'missings' in df.columns
```

Si volem eliminar totes les columnes que tingun algun valor `NaN`, ho indicarem amb el paràmetre `how=any`.

```python
df.dropna(axis=1, how='any').head()
```

El més habitual serà però un punt intermig. Mirarem quina és la disponibilitat de cadascuna de les variables, i decidirem si ens és útil per l'anàlisi o no. Amb la següent comanda, calculem el percentatge de valors `NaN` de cada variable.

```python
df.isnull().sum()/len(df)*100
```

Si establim com a criteri que no volem fer servir variables que tinguin més del 5% de valors `NaN`, eliminarem les variables `gross`, `content_rating`, `budget` i `aspect_ratio`. La funció `drop` ens permet eliminar les columnes pel seu nom.

```python
delete_columns = df.columns[df.isnull().sum()/len(df)*100 > 5]
delete_columns
```

```python
df.drop(delete_columns, axis=1).head()
```

### Eliminar les files amb variables no disponibles

En el nostre _dataset_ poden haver registres o files que no tinguin part de la informació. Una solució dràstica seria eliminar tots aquells registres que els falti alguna variable.

Tornarem a fer servir la funció `dropna` per aquest cas.

```python
df.dropna().head()
```

En l'altre extrem més conservador, podem eliminar aquelles files que tinguin tots els valors `NaN` amb el paràmetre `how=all`:

```python
df.dropna(how='all').head()
```

Podem afegir un _threshold_ en el nombre de valors `NaN` que estem disposats a tolerar en el nostre dataset. Si volem eliminar aquelles observacions amb 3 o més valors no disponibles, indicarem que volem totes les variables que tinguin com a mínim `len(df.columns)-3` valors no nuls:

```python
df.dropna(thresh=len(df.columns)-3).head()
```

Si hi ha alguna variable que necessitem que estigui informada, podem eliminar aquelles observacions que no la tinguin.

```python
df.dropna(subset=['director_name']).head()
```

### Omplir amb un valor per defecte

Una de les estratègies més habituals és omplir els valors no coneguts amb un valor per defecte. Per exemple, podriem decidir asumir que si no tenim dades dels _likes_ del director al _Facebook_, és que no en té cap.

```python
df.director_facebook_likes = df.director_facebook_likes.fillna(0)
```

O bé, podriem decidir calcular la mitjana dels _likes_ i imputar-li.

```python
df.director_facebook_likes = df.director_facebook_likes.fillna(df.director_facebook_likes.mean())
```

Les tècniques d'imputació que hem vist són les més senzilles de totes, però poden arribar a ser tan complexes com volguem.

## Transformació de les variables

Un pas previ imprescindible a aplicar un model a les dades és la realització d'algunes transformacions a les variables per adaptar-les a les característiques de l'anàlisi que volem realitzar. A continuació presentem tres de les transformacions més habituals.

### Conversió de tipus de dades

Hi ha vegades que els tipus de dades que infereix `Pandas` no són els que nosaltres volem. En aquest cas, haurem de convertir els tipus de dades de les variables. Per veure de quin tipus de dades és cada columna del nostre dataset, farem servir l'atribut `dtypes`.

```python
df.dtypes
```

Veiem que la variable `num_critic_for_reviews` és de tipus `float64`. Si volem convertir-la a enter, simplement farem servir la funció `astype('int')`. El tipus de dades `int` no accepta valors `NaN`, de manera que primer haurem d'assignar-li un valor a aquelles que no en tinguin. En el nostre cas, l'omplirem amb un 0.

```python
df.num_critic_for_reviews = df.num_critic_for_reviews.fillna(0)
df.num_critic_for_reviews.astype('int')
df.num_critic_for_reviews.head()
```

### Binarització de variables

La majoria d'eines i models per analitzar dades només accepten números com a valors de les variables. Això pot ser un problema si en les nostres variables tenim variables categòriques. Per convertir aquestes variables en numèriques, la manera més senzilla és fer el que es coneix per _one-hot encoding_, és a dir, transformar cada categoria de la variable en un vector binari que indiqui si la variable té aquest valor o no.

En el nostre dataset, per exemple, transformarem la variable _color_, que té dos valors diferents.

```python
df['color'].unique()
```

La funció `get_dummies` fa tota la feina del _one-hot encoding_.

```python
df_color = pd.get_dummies(df['color'])
df_color.head()
```

Ara només cal que eliminem la variable `color` en el nostre _dataset_ i la substituim per les dues variables binàries que hem creat.

```python
df.drop(['color'], axis = 1, inplace = True)
df = df.join(df_color)
df.head()
```
